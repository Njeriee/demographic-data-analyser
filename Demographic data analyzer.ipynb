```python
#minimum number of hours a person works per week
min_hours = demo['hours-per-week'].min()
min_hours
```

```python
#What percentage of the people who work the minimum number of hours per week have a salary of more than 50K
#total no of people working minimum hours
(demo['hours-per-week']==min_hours).sum()
#total number of people working minimum hours making more than 50k
((demo['hours-per-week']==min_hours) & (demo['salary']=='>50K')).sum()
#solution
(((demo['hours-per-week']==min_hours) & (demo['salary']=='>50K')).sum()/(demo['hours-per-week']==min_hours).sum())*100
```
